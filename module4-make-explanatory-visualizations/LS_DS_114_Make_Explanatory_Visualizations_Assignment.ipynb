```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Create data set to simulate graph
dems = ['Bullock','Biden', 'Klobuchar', 'Booker','Hickenlooper', 'Bennet', 'Harris','Gillibrand','Inslee', 'Castro','Sanders',"O'Rourke", 'Williamson','De Blasio','Warren', 'Yang','Gabbard', 'Delaney', 'Buttigieg','Ryan','Gravel', 'Swallwell', 'Moulton']
total_emails = np.array([59, 65, 45, 68, 45, 54, 36, 48, 30, 46, 42, 38, 30, 8, 56, 24,12, 20, 16, 78, 5,2,0 ])
trump_mentions = np.array([59,27, 18, 16, 16, 14, 14, 13, 13, 12, 12, 11, 8, 6, 5, 5, 5, 2, 2, 0, 0,0,0])

# Access figure and axes objects
fig, ax = plt.subplots()

# Set figure height and width, in inches
fig.set_figheight(8)
fig.set_figwidth(5)

# Create y tick marks
spc = 75
idx = np.arange(0, spc*len(dems), spc)

# Flip the y axis so that candidate 0/'Bullock' shows up on top 
ax.invert_yaxis()

# Get the xaxis ticks to to show up on top of the graph
ax.xaxis.tick_top()

# Don't display the y grid lines
ax.grid(b=False, axis='y')

# Set the title
ax.set_title('Which candidates are talking about Trump?', fontweight='bold', loc='left', fontsize=12, x=-0.2, y=1.1)

# Set the subtitle
ax.text(x = -15,
        y = -200,
        s='Number of emails FiveThirtyEight received from 2020 Democratic\ncandidates and how many of those emails contain the word "Trump"\n',
        fontsize=11
        )

# Set y ticks every 75 places on the y axis - 0, 75, 150, ...
ax.set_yticks(idx)
# Label them with Democratic candidate names
ax.set_yticklabels(dems, fontsize=9)

# Set the x lims so that that end grid lines are drawn
ax.set_xlim(-1, 81)

# Set x ticks every 20 places on the y axis - 0, 20, 40, ... 80
ax.set_xticks(range(0,100,20))
# Set the tick labels color gray
ax.set_xticklabels(labels=range(0,100, 20), color='gray', fontsize=8)

# Draw the 23 bars denoting the number of emails mentioning trump
ax.barh(idx, trump_mentions, color='#FC7A64', height=70);
# Stack the 23 bars denoting the number of emails not mentioning trump above those
ax.barh(idx, total_emails-trump_mentions, left=trump_mentions, color='#FEA08B', height=70)

#Each of the 46 bars drawn are added to ax.patches

# The Biden mentions and doesn't mention Trump bars are patches 1 and 24(1+23), respectively

p1 = ax.patches[1]
p2 = ax.patches[24]

# Get the x coords for the two text boxes
x1 = p1.get_width()-24
x2 = x1 + p2.get_width() - 8
# Compute the y coord for the text boxes
y = p1.get_y()+45

# Draw the text boxes
ax.text(x1, y,'Mentions Trump', color='white', fontsize=9)

ax.text(x2, y,"Doesn't mention Trump", color='white', fontsize=9);
```
