```python
# Clean up the frame by skipping the first row

df = pd.read_csv('LoanStats_2018Q4.csv', skiprows=1)

display(df.shape)
df.head()
```

```python
# Clean up the frame by saying that the header row is the second row

df = pd.read_csv('LoanStats_2018Q4.csv', header=1)

display(df.shape)
display(df.head())
display(df.tail())
```

```python
pd.set_option('display.max_rows', 200)

#Sum null values by column and display in ascending order
df.isnull().sum().sort_values()
```

```python
# Every column has at least 2 NaNs from the last 2 lines
# Fix the problem by skipping the last 2 lines
# We could use the skipfooter option in read_csv but
# that results the slower python engine being used.

df = pd.read_csv('LoanStats_2018Q4.csv', header=1)[:-2]

display(df.shape)

display(df.head())
display(df.tail())

```

```python
#Sum null values by column and display in ascending order
df.isnull().sum().sort_values()
```

```python
#determins empty columna

empty_columns = [col for col in df.columns if  df[col].isnull().sum() == len(df[col])]

empty_columns
```

```python
#Drop the empty columns

df = df.drop(empty_columns, axis=1)

df.head()
```

```python
# Examine int_rate values
display(df.int_rate.head())
display(df.int_rate[3])
```

```python
# int_rate column has string values with a leading space and trailing '%'
# Strip the leading space and trailing %, then cast the string to a float
df['int_rate_cleaned'] = df.int_rate.apply(lambda x: float(x.strip().strip('%')))
display(df.dtypes[-5:])
```

```python
display(df.int_rate_cleaned)
```

```python
cols = [col for col in df if col.endswith('_d')]
cols
```

```python
display(df.issue_d[:10])
display(df.issue_d[9])
```

```python
df.issue_d = pd.to_datetime(df.issue_d, infer_datetime_format=True)

#Verify that issue_d has changed to datetime
df.dtypes[:15]
```

```python
#Check an issue_d value to see how datetime is represented

df.issue_d.iloc[0]
```

```python
#Add 2 new cols for the year and month of issue

df['issue_year'] = df.issue_d.dt.year
df['issue_month'] = df.issue_d.dt.month

df.head()
```

```python
#Check the dtypes for the 2 new columns

df.dtypes[-2:]
```

```python
df.earliest_cr_line.head()
```

```python
df.earliest_cr_line = pd.to_datetime(df.earliest_cr_line, infer_datetime_format=True)
```

```python
#Create a new column that displays the number of days from earliest credit to issue

df['days_from_earliest_credit_to_issue'] = (df.issue_d - df.earliest_cr_line).dt.days

df.head()
```

```python
# Convert term values to int
df.term = df.term.apply(lambda x: int(x.strip().split()[0]))

df.term.head()
```

```python
df['loan_status_is_great'] = df.loan_status.apply(lambda x : int(x in ['Current', 'Fully Paid']))

df.loan_status_is_great.head(10)
```

```python
# Create last_pymnt_d_month by converting last_pymnt_d to datatime and extracting the month
df['last_pymnt_d_month'] = (pd.to_datetime(df.last_pymnt_d, infer_datetime_format=True)).dt.month
display(df.last_pymnt_d_month.head(10))
display(df.last_pymnt_d.head(10))

# Some rows have no value for 'last_pyment_d' i.e. Nan - therefore the value for 'last_pymnt_d_month' becomes a float
display(df.last_pymnt_d.isnull().sum())
```

```python
# Create last_pymnt_d_month by converting last_pymnt_d to datatime and extracting the month
df['last_pymnt_d_year'] = (pd.to_datetime(df.last_pymnt_d, infer_datetime_format=True)).dt.year
display(df.last_pymnt_d_year.head(10))

#As with last_pymnt_d_month, last_pymnt_d_year is of type float
```
